```python
#task 4
import pandas as pd
import numpy as np
 
# Load the CSV files into DataFrames
project_df = pd.read_csv('project.csv')
employee_df = pd.read_csv('employee.csv')
seniority_df = pd.read_csv('seniority.csv')
 
# Merge all three DataFrames on the 'ID' column
# First, merge project_df with employee_df
merged_df = pd.merge(project_df, employee_df, on='ID', how='inner')
 
# Then, merge the result with seniority_df
final_df = pd.merge(merged_df, seniority_df, on='ID', how='inner')
 
# Display the final DataFrame
print("Final DataFrame:")
print(final_df)
 
# Save the final DataFrame to a CSV file (optional)
final_df.to_csv('final.csv', index=False)
```

```python
import pandas as pd
 
# Load the Final DataFrame from the CSV file
final_df = pd.read_csv('final.csv')
 
# Add a new 'Bonus' column initialized to 0
final_df['Bonus'] = 0
 
# Calculate the bonus only for finished projects
final_df.loc[final_df['Status'] == 'Finished', 'Bonus'] = final_df['Cost'] * 0.05
 
# Save the updated DataFrame to the same CSV file
final_df.to_csv('final.csv', index=False)
 
# Display the updated DataFrame
print("Updated Final DataFrame with Bonus:")
print(final_df)
```

```python
import pandas as pd
 
# Load the Final DataFrame from the CSV file
final_df = pd.read_csv('final.csv')
 
# Load the Seniority DataFrame to ensure we have the correct designation levels
seniority_df = pd.read_csv('seniority.csv')
 
# Merge the Final DataFrame with Seniority DataFrame to include the correct 'Designation Level'
final_df = final_df.merge(seniority_df, on='ID', how='left', suffixes=('', '_seniority'))
 
# Update the 'Designation Level' for employees with failed projects
# Use the 'Designation Level_seniority' column for updating
final_df.loc[final_df['Status'] == 'Failed', 'Designation Level'] = final_df.loc[final_df['Status'] == 'Failed', 'Designation Level_seniority'] - 1
 
# Remove records where designation level is above 4
final_df = final_df[final_df['Designation Level'] <= 4]
 
# Drop unnecessary columns
final_df.drop(columns=['Designation Level_seniority'], inplace=True)
 
# Save the updated DataFrame back to the existing CSV file
final_df.to_csv('final.csv', index=False)
 
# Display the updated DataFrame
print("Updated Final DataFrame:")
print(final_df)
```

```python
import pandas as pd

# Load the final.csv file into a DataFrame
final_df = pd.read_csv('final.csv')

# Define a function to add "Mr." or "Mrs." based on the Gender
def add_prefix(row):
    if row['Gender'] == 'M':
        return 'Mr. ' + row['First Name']
    elif row['Gender'] == 'F':
        return 'Mrs. ' + row['First Name']
    return row['First Name']

# Apply the function to the 'First Name' column
final_df['First Name'] = final_df.apply(add_prefix, axis=1)

# Drop the 'Gender' column
final_df.drop(columns=['Gender'], inplace=True)

# Save the modified DataFrame back to the same CSV file
final_df.to_csv('final.csv', index=False)

print(final_df)
```

```python
import pandas as pd

# Load the final.csv file into a DataFrame
final_df = pd.read_csv('final.csv')

# Define a function to promote designation level
def promote_designation(row):
    if row['Age'] > 29:
        return row['Designation Level'] + 1
    return row['Designation Level']

# Apply the function to the 'Designation Level' column
final_df['Designation Level'] = final_df.apply(promote_designation, axis=1)

# Save the modified DataFrame back to the same CSV file
final_df.to_csv('final.csv', index=False)

print(final_df)
```
